```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

#      
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

#  
G = nx.Graph()

#   (, , , )
jobs = [" -  ", "  ", "  ", " "]
skills = ["", "SAS", "", "QGIS", "", ""]
projects = [" ", " ", " "]
industries = ["", "", " ", ""]

#   
for job in jobs:
    G.add_node(job, label="Job")

#   
for skill in skills:
    G.add_node(skill, label="Skill")

#   
for project in projects:
    G.add_node(project, label="Project")

#   
for industry in industries:
    G.add_node(industry, label="Industry")

# -   ( )
G.add_edges_from([
    (" -  ", "SAS"),
    (" -  ", ""),
    ("  ", ""),
    ("  ", ""),
    ("  ", "QGIS"),
    (" ", ""),
    (" ", "")
])

# -   ( )
G.add_edges_from([
    ("  ", " "),
    ("  ", " "),
    (" ", " ")
])

# -   ( )
G.add_edges_from([
    (" -  ", ""),
    ("  ", " "),
    ("  ", ""),
    (" ", "")
])

#  
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightgreen", font_size=10, node_size=3000, font_weight="bold")
plt.title("    ", fontsize=15)
plt.show()
```
